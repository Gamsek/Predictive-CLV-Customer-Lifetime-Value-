## **Business Problem Understanding**

**Context**

Customer Lifetime Value (CLV) adalah metrik yang mengukur nilai seorang pelanggan bagi perusahaan sepanjang masa hubungan mereka. CLV membantu perusahaan memahami jumlah keuntungan yang dihasilkan dari satu pelanggan serta biaya yang diperlukan untuk mendapatkan atau mempertahankan pelanggan tersebut. Mengetahui CLV sangat penting untuk menentukan strategi pemasaran yang efektif dan memprediksi perubahan perilaku pelanggan di masa depan. 

Dalam konteks perusahaan asuransi, mengoptimalkan biaya akuisisi pelanggan adalah kunci untuk memaksimalkan CLV. Perusahaan asuransi menawarkan berbagai jenis produk asuransi kendaraan dengan berbagai tingkat cakupan dan premi. Setiap pelanggan memiliki karakteristik unik seperti kelas kendaraan, jenis asuransi, status pekerjaan, status pernikahan, tingkat pendidikan, jumlah polis, premi bulanan, jumlah klaim total, dan pendapatan tahunan. Data ini sangat berharga untuk memahami profil pelanggan dan memprediksi nilai seumur hidup mereka. [Menurut McKinsey & Company](https://www.mckinsey.com/capabilities/mckinsey-digital/our-insights/customer-lifetime-value-the-customer-compass), perusahaan yang menggunakan analisis data untuk memahami CLV dapat merancang strategi pemasaran yang lebih efektif dan mengurangi biaya dari pemasaran itu sendiri.

Dengan memprediksi CLV, perusahaan dapat menargetkan pelanggan yang paling bernilai dan mengoptimalkan strategi pemasaran. Misalnya, pelanggan dengan CLV tinggi dapat diberikan penawaran khusus atau layanan yang lebih personal untuk meningkatkan loyalitas mereka. Sebaliknya, pelanggan dengan CLV rendah dapat dianalisis lebih lanjut untuk menentukan apakah ada cara yang efektif untuk meningkatkan nilai mereka atau memutuskan apakah pengeluaran lebih lanjut untuk mereka tidak memberikan ROI yang cukup tinggi. Penerapan model prediksi CLV ini dapat mengurangi biaya pemasaran dan meningkatkan ROI pemasaran. 

**Problem Statement**

Tim marketing dari perusahaan asuransi menghadapi masalah dalam **mengoptimalkan strategi pemasaran karena kurangnya prediksi yang akurat tentang untuk customer yang memiliki nilai yang menguntungkan bagi perusahaan**. Tim marketing kesulitan menentukan alokasi anggaran pemasaran dan retensi yang tepat karena tidak mengetahui nilai seumur hidup pelanggan secara akurat. CFO mengalami tantangan dalam mengelola anggaran secara efisien akibat pengeluaran yang tidak tepat sasaran. Masalah ini mengakibatkan tingginya biaya akuisisi pelanggan baru tanpa peningkatan signifikan dalam nilai seumur hidup pelanggan. Penyelesaian masalah ini penting **untuk meningkatkan profitabilitas perusahaan dengan mengurangi biaya pemasaran dan meningkatkan return on investment (ROI) pemasaran**.

**Goals**

Tujuan dari penyelesaian masalah ini adalah **mengembangkan model machine learning yang dapat memprediksi Customer Lifetime Value (CLV) dengan akurasi tinggi**. Target yang ingin dicapai adalah mengurangi biaya akuisisi pelanggan baru serta meningkatkan return on investment (ROI) pemasaran. Dengan pencapaian target ini, perusahaan dapat mengalokasikan anggaran pemasaran secara lebih efisien dan fokus pada pelanggan yang memiliki nilai seumur hidup tinggi, sehingga meningkatkan profitabilitas keseluruhan.

**Analytic Approach**

Rancangan solusi yang ditawarkan untuk menyelesaikan masalah ini adalah membangun model regresi machine learning untuk memprediksi CLV berdasarkan data historis pelanggan. Model ini akan dilatih menggunakan berbagai fitur pelanggan seperti kelas kendaraan, jenis asuransi, status pekerjaan, status pernikahan, tingkat pendidikan, jumlah polis, premi bulanan, jumlah klaim total, dan pendapatan. User akan menggunakan solusi ini setiap kali ada pelanggan baru atau pembaruan data pelanggan yang ada. Tim Marketing dapat menggunakan prediksi CLV untuk menargetkan pelanggan bernilai tinggi dalam kampanye pemasarannya, sementara CFO dapat menggunakan prediksi ini untuk mengalokasikan anggaran pemasaran dengan lebih efisien.

**Metric Evaluation**

Kondisi miss-prediction terjadi jika model memprediksi CLV dengan kesalahan yang signifikan, yang mengakibatkan alokasi anggaran pemasaran dan retensi yang tidak efisien. Konsekuensinya adalah pemborosan sumber daya dan kurangnya fokus pada pelanggan yang sebenarnya bernilai tinggi. Metrik evaluasi yang digunakan untuk mengukur keberhasilan adalah Mean Absolute Error (MAE), Mean Squared Error (MSE), Root Mean Squared Error (RMSE), Mean Absolute Percentage Error (MAPE), dan R-squared (R²). MAE mengukur rata-rata kesalahan absolut antara prediksi dan nilai aktual, dengan nilai yang lebih rendah menunjukkan prediksi yang lebih akurat. MSE mengukur rata-rata kesalahan kuadrat antara prediksi dan nilai aktual, dengan nilai yang lebih rendah juga menunjukkan prediksi yang lebih akurat. RMSE, sebagai akar dari MSE, memberikan bobot lebih pada kesalahan yang lebih besar. MAPE mengukur rata-rata persentase kesalahan absolut, memberikan interpretasi yang lebih intuitif. R-squared (R²) mengukur proporsi variansi dalam data yang dapat dijelaskan oleh model, dengan nilai yang lebih tinggi menunjukkan model yang lebih baik. Metrik evaluasi ini membantu mengevaluasi seberapa akurat model dalam memprediksi CLV, sehingga strategi pemasaran dapat dioptimalkan sesuai dengan tujuan bisnis. Asumsi yang digunakan adalah data historis pelanggan yang tersedia cukup representatif untuk membangun model prediksi yang akurat dan dapat menggeneralisasi dengan baik ke data pelanggan baru.

**Korelasi Data pada Dataset CLV**

Untuk melihat korelasi antara fitur-fitur, kita akan menghitung matriks korelasi dan membuat heatmap. 

```python
numerical_features = clv.select_dtypes(include=['float64', 'int64']).columns

# Menghitung matriks korelasi menggunakan metode Spearman (karena data tidak distribusi normal)
corr_matrix_spearman = clv[numerical_features].corr(method='spearman')

# Membuat heatmap dari matriks korelasi
plt.figure(figsize=(12, 8))
matrix = np.triu(corr_matrix_spearman)
sns.heatmap(corr_matrix_spearman, annot=True, fmt=".2f", mask=matrix ,  cmap='magma')
plt.title('Heatmap Korelasi Fitur')
plt.show()
```

Saya akan menampilkan data yang memiliki outlier dengan menggunakan dataframe.

```python
# Fungsi untuk menghitung dan mengidentifikasi outlier berdasarkan IQR
def calculate_iqr_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Mengidentifikasi outlier
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
    
    # Menghitung jumlah data di bawah lower bound dan di atas upper bound
    below_lower_bound = data[data[column] < lower_bound].shape[0]
    above_upper_bound = data[data[column] > upper_bound].shape[0]
    
    return lower_bound, upper_bound, len(outliers), below_lower_bound, above_upper_bound

# Membuat dictionary untuk menyimpan hasil
hasil_outlier = {}

# Melakukan perhitungan outlier untuk semua kolom numerik
for column in numerical_features:
    lb, ub, outliers_count, below_lb, above_ub = calculate_iqr_outliers(clv_model, column)
    hasil_outlier[column] = {
        "Lower Bound": lb,
        "Upper Bound": ub,
        "Jumlah di Bawah Lower Bound": below_lb,
        "Jumlah di Atas Upper Bound": above_ub,
        "Jumlah Outliers": outliers_count
    }

# Membuat DataFrame dari hasil outlier
outlier_df = pd.DataFrame(hasil_outlier).transpose()
outlier_df

# Menghitung total outliers dari setiap kolom
# Menghitung total outliers dari setiap kolom
total_row = pd.DataFrame(outlier_df.sum(axis=0)).transpose()
total_row.index = ['Total']

# Menambahkan baris total ke DataFrame
outlier_df = pd.concat([outlier_df, total_row])
outlier_df
```

Teknik ini menjaga informasi penting yang mungkin hilang jika outliers dihapus sepenuhnya. **Membuat Pipeline**

```python
import numpy as np
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from scipy.stats import mstats
import pandas as pd

# Membuat class untuk menangani outlier
class IQRHandler(BaseEstimator, TransformerMixin):
    def __init__(self, factor=1.5):
        self.factor = factor

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        for col in X.select_dtypes(include=[np.number]).columns:
            Q1 = X[col].quantile(0.25)
            Q3 = X[col].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - (self.factor * IQR)
            upper_bound = Q3 + (self.factor * IQR)
            X[col] = X[col].apply(lambda x: lower_bound if x < lower_bound else (upper_bound if x > upper_bound else x))
        return X
    
    def get_feature_names_out(self, input_features=None):
        return input_features

class Winsorizer(BaseEstimator, TransformerMixin):
    def __init__(self, limits=(0.05, 0.05)):
        self.limits = limits

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        X_winsorized = X.select_dtypes(include=[np.number]).apply(lambda col: mstats.winsorize(col, limits=self.limits))
        X[X_winsorized.columns] = X_winsorized
        return X
    
    def get_feature_names_out(self, input_features=None):
        return input_features

# Membuat class untuk menangani missing value 
class MissingHandler(BaseEstimator, TransformerMixin):
    def __init__(self, fill_value=1):
        self.fill_value = fill_value

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        X['Policy Duration (Months)'] = X['Policy Duration (Months)'].replace(0, self.fill_value)
        return X

    def get_feature_names_out(self, input_features=None):
        return input_features

# Membuat custom ordinal encoder
class CustomOrdinalEncoder(BaseEstimator, TransformerMixin):
    def __init__(self, ordinal_mapping):
        self.ordinal_mapping = ordinal_mapping

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()
        for mapping in self.ordinal_mapping:
            col = mapping['col']
            X[col] = X[col].map(mapping['mapping'])
        return X

    def get_feature_names_out(self, input_features=None):
        return input_features

# Membuat ordinal mapping untuk kolom data yang sudah ditentukan
ordinal_mapping = [
    {'col':'Coverage', 'mapping':{'Basic':1, 'Extended':2, 'Premium':3, None:0}},
    {'col':'Education', 'mapping':{'High School or Below':1, 'College':2, 'Bachelor':3, 'Master':4, 'Doctor':5, None:0}},
    {'col': 'Vehicle Class', 'mapping':{"Four-Door Car":1, "Two-Door Car":2, "SUV":3, "Sports Car":4, "Luxury SUV":5, "Luxury Car":6, None:0}},
    {'col': "Renew Offer Type", 'mapping': {"Offer1": 1, "Offer2":2, "Offer3":3, 'Offer4':4, None:0}},
    {'col': "Location Code", 'mapping': {'Rural':1, 'Suburban':2, 'Urban':3, None:0}}
]

# Mendefinisikan kolom sesuai dengan metode masing-masing
ordinal_cols = [mapping['col'] for mapping in ordinal_mapping]
onehot_cols = ["Gender", "Marital Status", "Employment Status"]

# Membuat step preprocessing encode data kategorikal
preprocessor = ColumnTransformer(
    transformers=[
        ('ordinal', CustomOrdinalEncoder(ordinal_mapping), ordinal_cols),
        ('onehot', OneHotEncoder(), onehot_cols)
    ],
    remainder='passthrough'
)

pipeline_c = Pipeline(steps=[
    ('iqr', IQRHandler(factor=1.5)), 
    ('winsorize', Winsorizer(limits=(0.05, 0.05))),  
    ('fill_policy_duration', MissingHandler(fill_value=1)),
    ('preprocessor', preprocessor)
])
```

### **Rekomendasi**

#### 1. Projek Secara Umum

**Data Collection dan Quality:**
- **Perbaikan Kualitas Data:** Pastikan data yang dikumpulkan bersih, lengkap, dan akurat. Lakukan preprocessing yang lebih teliti untuk menangani missing values, outliers, dan inkonsistensi dalam data.
- **Frekuensi Pengumpulan Data:** Perbarui dataset secara berkala untuk menangkap perubahan perilaku pelanggan dan dinamika pasar. Pengumpulan data yang lebih sering dapat membantu model tetap relevan dan akurat.
- **Ekspansi Variabel:** Pertimbangkan untuk menambahkan lebih banyak variabel yang relevan, seperti data demografi, perilaku pembelian, interaksi pelanggan dengan layanan, dan data sentimen dari ulasan pelanggan. Variabel tambahan ini dapat memberikan lebih banyak informasi bagi model untuk membuat prediksi yang lebih akurat.

**Waktu dan Sumber Daya:**
- **Penjadwalan Iterasi Model:** Tetapkan siklus iterasi model yang jelas, misalnya setiap tiga bulan, untuk menguji, memvalidasi, dan memperbarui model berdasarkan data terbaru.
- **Alokasi Sumber Daya:** Pastikan tim memiliki akses ke sumber daya yang cukup, termasuk perangkat keras untuk komputasi, perangkat lunak untuk analisis, dan tenaga ahli yang mampu menjalankan analisis data dan pengembangan model.

#### 2. Proses Pengembangan Model

**Model Tuning dan Validation:**
- **Hyperparameter Tuning:** Lakukan tuning hyperparameter yang lebih komprehensif menggunakan teknik seperti Grid Search atau Random Search untuk menemukan kombinasi parameter yang optimal bagi model Random Forest Regressor.
- **Cross-Validation:** Gunakan teknik cross-validation yang lebih ekstensif seperti k-fold cross-validation untuk memastikan bahwa model tidak overfit pada data training dan dapat melakukan generalisasi dengan baik pada data unseen.

**Ensembling dan Model Comparison:**
- **Teknik Ensembling:** Pertimbangkan untuk menggunakan teknik ensembling lain seperti Gradient Boosting atau AdaBoost, dan bandingkan kinerjanya dengan model Random Forest. Teknik ensembling lain dapat membantu mengurangi kesalahan prediksi dan meningkatkan akurasi.
- **Model Stacking:** Uji pendekatan model stacking, di mana beberapa model digabungkan untuk meningkatkan prediksi akhir. Stacking memungkinkan penggunaan kekuatan dari berbagai model untuk membuat prediksi yang lebih akurat.

**Feature Engineering:**
- **Pembuatan Fitur Baru:** Identifikasi dan buat fitur baru yang dapat meningkatkan kinerja model. Fitur interaksi, transformasi logaritmik untuk variabel yang sangat skewed, dan fitur waktu (seperti musim atau tren bulanan) dapat memberikan nilai tambah.
- **Seleksi Fitur:** Gunakan teknik seleksi fitur seperti Recursive Feature Elimination (RFE) untuk mengidentifikasi dan mempertahankan fitur-fitur yang paling berpengaruh terhadap prediksi CLV.

**Monitoring dan Maintenance:**
- **Pemantauan Kinerja Model:** Implementasikan sistem pemantauan kinerja model secara real-time untuk mendeteksi degradasi performa. Buat metrik yang jelas untuk evaluasi berkala model setelah implementasi.
- **Pembaruan Model:** Rencanakan pembaruan model secara berkala berdasarkan data terbaru dan evaluasi performa model. Ini akan memastikan bahwa model tetap akurat dan relevan dengan kondisi pasar dan perilaku pelanggan yang berubah.

Dengan mengikuti rekomendasi ini, proses pengembangan model prediksi CLV akan lebih terstruktur dan efisien, serta memberikan hasil yang lebih akurat dan dapat diandalkan. Ini akan membantu bisnis dalam membuat keputusan yang lebih tepat dan mengoptimalkan strategi pemasaran dan retensi pelanggan.
